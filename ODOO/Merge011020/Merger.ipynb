```python
inv = pd.read_excel("RGM2020x.xlsx")
```

```python
inv.reset_index(drop=True,inplace=True)
```

### Nettoyer le Dataframe

```python
inv = inv.drop(['Stock Comptable','Montant Comptable','Unnamed: 10'],axis=1)
```

### Extraire Famille

```python
#Famille
t_famille=[]
c=0
for i in inv['Symbole']:
    print(i)
    c = c+1
    
    s = ''
    x=0
    for j in i:        
        if not j.isnumeric():            
            s = j
            break
            print(s)
       
    print(s)
    t_famille.append(s)
    continue
    
    #if c >20:
    #    break
    
print('\n\n ## ## Done ## ## ')
```

```python
len(t_famille)
```

### Extraire Sous Famille

```python
#Sous Famille
t_sous_famille =[]
c=0
for i in inv['Symbole']:
    print(i)
    c = c+1
    
    s = ''
    for j in i:
        if not j.isnumeric():
            s = s+j
    print(s)
    t_sous_famille.append(s)
    
    #if c >20:
    #    break
    
print('\n\n ## ## Done ## ## ')
```

```python
len(t_sous_famille)
```

### Combiner 

```python
inv["code famille"] = t_famille
```

```python
inv["code sous famille"] = t_sous_famille
```
